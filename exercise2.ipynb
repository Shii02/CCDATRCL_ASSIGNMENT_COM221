```python
# Creating a stack
def create_stack():
    stack = []
    return stack

def print_stack(stack):

    if (check_empty(stack)):
            return "stack is empty"

    # Create a copy of the stack and reverse the elements
    tempStack = stack.copy()
    tempStack.reverse()

    # Print all elements using a loop
    for i in tempStack:
        print(i)

    # Delete the copy to free up memory
    del tempStack

# Creating an empty stack
def check_empty(stack):
    return len(stack) == 0


# Adding items into the stack
def push(stack, item):
    stack.append(item)
    print("pushed item: ", item)


# Removing an element from the stack
def pop(stack):
    if (check_empty(stack)):
        return "stack is empty"

    return stack.pop()

def peek(stack):

    # Use negative indexing to display the last element of the list
    print(stack[-1])
```

```python
# item no 1 : create a burger stack

stack = create_stack()

push(stack, "Bun")
push(stack, "Patty")
push(stack, "Cheese")
push(stack, "Lettuce")
push(stack, "Tomato")
push(stack, "Bun")

print("\n---\n")
print_stack(stack)
```

```python
# item no 2 : reverse a string using a stack

name_stack = create_stack()

push(name_stack, 'C')
push(name_stack, 'A')
push(name_stack, 'R')
push(name_stack, 'L')
push(name_stack, 'A')
push(name_stack, 'R')
push(name_stack, 'V')
push(name_stack, 'I')
push(name_stack, 'N')

print(name_stack, "\n")

reversed_stack = create_stack()

while len(name_stack) > 0:
  push(reversed_stack, pop(name_stack))

print(reversed_stack)
```

```python
# item no 3 : remove elements using a stack

number_stack = create_stack()

for i in range(1, 11):
  push(number_stack, i)

print(number_stack, "\n")

eNumber_stack = create_stack()

while number_stack:
  num = pop(number_stack)

  if num % 2 == 0:
    push(eNumber_stack, num)

eNumber_stack.reverse()
print("\nEven Numbers :", eNumber_stack)
```

```python
# item no 4 : inserting elements into a stack

name_stack = create_stack()

push(name_stack, 'C')
push(name_stack, 'A')
push(name_stack, 'R')
push(name_stack, 'L')
push(name_stack, 'H')
push(name_stack, 'I')
push(name_stack, 'P')
push(name_stack, 'O')
push(name_stack, 'L')
push(name_stack, 'I')
push(name_stack, 'T')
push(name_stack, 'O')

print(name_stack, "\n")

empty_stack = create_stack()

while name_stack:
  str1 = pop(name_stack)

  if str1 in ['A', 'E', 'I', 'O', 'U']:
    if str1 not in empty_stack:
      push(empty_stack, str1)

empty_stack.reverse()
print(empty_stack)

```

```python
# item no 5 : sort elements in a stack

number_stack = create_stack()

push(number_stack, 3)
push(number_stack, 18)
push(number_stack, 7)
push(number_stack, 10)
push(number_stack, 12)
push(number_stack, 5)



print("Numbers :", number_stack, "\n")

sorted_stack = create_stack()

def display():
  print("Number Stack :", number_stack)
  print("Sorted Stack :", sorted_stack)
  print("---")

def sortNumber():
  num1 = pop(number_stack)

  if len(sorted_stack) > 0:
    num2 = pop(sorted_stack)

    if num1 >= num2:
      push(sorted_stack, num2)
      push(sorted_stack, num1)
      display()

    else:
      push(sorted_stack, num1)
      push(number_stack, num2)
      display()

      stop = 0

      while stop == 0:
        if len(sorted_stack) >= 2:
          num1 = pop(sorted_stack)
          num2 = pop(sorted_stack)

          if num1 >= num2:
            push(sorted_stack, num2)
            push(sorted_stack, num1)
            stop = 1

          else:
            push(number_stack, num2)
            push(sorted_stack, num1)
            display()
        else:
          break;

  else:
    push(sorted_stack, num1)
    display()


while len(number_stack) > 0:
  sortNumber()

print("Sorted Numbers :", sorted_stack)
```
